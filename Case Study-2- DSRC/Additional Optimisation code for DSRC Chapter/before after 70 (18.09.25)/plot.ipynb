```python
import pandas as pd
import matplotlib.pyplot as plt

# Correct file names (no .csv extension in your case)
before_file = "overall_metrics_summary_Jerk_70.csv"
after_file = "each_metrics_summary_Jerk_70.csv"

# Load CSVs (force .csv engine even if no extension)
before_df = pd.read_csv(before_file)
after_df = pd.read_csv(after_file)

LOS_list = ['A', 'B', 'C', 'D', 'E']

fig, axes = plt.subplots(len(LOS_list), 2, figsize=(10, 12), sharex=True, sharey=True)
fig.suptitle("Histogram of Anomaly_Prob Before vs After", fontsize=14)

for i, los in enumerate(LOS_list):
    before_vals = before_df.loc[before_df['LOS'] == los, 'Anomaly_Prob'].dropna()
    after_vals = after_df.loc[after_df['LOS'] == los, 'Anomaly_Prob'].dropna()

    axes[i, 0].hist(before_vals, bins=20, edgecolor='black')
    axes[i, 0].set_title(f"LOS {los} - Before")
    axes[i, 0].set_ylabel("Count")

    axes[i, 1].hist(after_vals, bins=20, edgecolor='black')
    axes[i, 1].set_title(f"LOS {los} - After")

for ax in axes[-1, :]:
    ax.set_xlabel("Anomaly_Prob")

plt.tight_layout(rect=[0, 0, 1, 0.97])
plt.show()
```
