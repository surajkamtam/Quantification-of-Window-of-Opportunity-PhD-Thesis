```python
import pandas as pd

data = pd.read_csv('Jan60Nresults.csv')
df = pd.DataFrame(data)

# Reshape (melt) the DataFrame to have Date, Time, and LOS in separate columns
df_melted = pd.melt(df, id_vars=["Local Date"], var_name="Time", value_name="LOS")

# Show the first few rows of the reshaped data
df_melted.head()
```

```python
output_file_path = 'df_melted_results.csv'
df_melted.to_csv(output_file_path, index=False)
```
